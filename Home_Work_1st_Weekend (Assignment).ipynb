## **1.** Identify columns with more than **20% missing values** and drop them


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('melb_data.csv')
df.head()
```

```python
df.info()
```

```python
missing_person = df.isna().sum()
print(missing_person)

missing_person_percentage = missing_person/len(df)*100
print(missing_person_percentage)

Column_to_drop = missing_person_percentage[missing_person_percentage>20].keys()
print(Column_to_drop)

df_cleaned = df.drop(columns = Column_to_drop)
print(df_cleaned.info())
df_cleaned.head()
```

## **2.**	Fill missing values in the **numerical columns** using **median instead of mean.**




```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Check which numeric columns have missing values
missing_numeric_data = df[numeric_cols].isna().sum()

# Display the numeric columns with missing data and their count
print(missing_numeric_data[missing_numeric_data > 0])
```

```python
df.loc[:, "Car"] = df["Car"].fillna(df["Car"].median())
df.loc[:, "BuildingArea"] = df["BuildingArea"].fillna(df["BuildingArea"].median())
df.loc[:, "YearBuilt"] = df["YearBuilt"].fillna(df["YearBuilt"].median())
df.head()
```

# **3.**	Fill missing values in the **categorical columns** using the most frequent value **(mode)**.

```python
df["Suburb"].fillna(df["Suburb"].mode()[0], inplace=True)
df["Type"].fillna(df["Type"].mode()[0], inplace=True)

print(df)
```
